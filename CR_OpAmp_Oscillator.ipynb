```python
f0 = 16E+3
fmin = f0-0.5E+3
fmax = f0+0.5E+3
C1 = 280E-12

fs  = np.array([fmin, f0, fmax])
R3s = 1/(2*np.log(2)*C1*fs)

for f, R3 in zip(fs, R3s):
    print(f"f={f:.2e} Hz, R3={R3/1000:.2f} kOhm")
    
print((R3s.max()-R3s.min())/1000)
```

```python
R1 = 10E+3
R2 = 10E+3
R3v = np.arange(0, 10.1E+3, 100) # 可変抵抗の範囲

#R3o 固定抵抗の値

components = [
    {"R3o": 156.14E+3, "C1": 280E-12},
    {"R3o": 145.43E+3, "C1": 320E-12},
    {"R3o": 134.45E+3, "C1": 370E-12},
    {"R3o": 124.26E+3, "C1": 430E-12},
    {"R3o": 115.42E+3, "C1": 500E-12},
    {"R3o": 104.54E+3, "C1": 600E-12},
    ]

plt.figure(figsize=(8,10))

for comp in components:
    R3o = comp["R3o"]
    C1 = comp["C1"]
    f = 1/(2*(R3o+R3v)*C1*np.log(1+R1/R2))

    plt.plot((R3v)/1000, f/1000, label=f'R3o={R3o/1000:.1f} kOhm, C1={C1/1E-12:.0f} pF, f = {f.min()/1000:.2f}-{f.max()/1000:.2f} kHz')

plt.legend()
    

plt.xlabel('R3 (kOhm)')
plt.ylabel('f (kHz)')
plt.ylim(10, 20)
plt.yticks(np.arange(10, 21, 1))
plt.grid()
```
