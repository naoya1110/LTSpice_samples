$f_0 = \frac{1}{2R_3C_1\ln(1+R1/R2)}$

```python
R1 = 10E+3
R2 = 10E+3
R3o = 20E+3
R3v = np.arange(0, 10.1E+3, 100) # 可変抵抗の範囲
C1 = 300E-12

plt.figure(figsize=(10,10))

for R3o in [55E+3, 70E+3, 85E+3, 100E+3]:
    f = 1/(2*np.pi*(R3o+R3v)*C1*np.log(1+R1/R2))

    plt.plot((R3v)/1000, f/1000, label=f'R3o = {(R3o/1000):.0f} kOhm, f = {f.min()/1000:.2f}-{f.max()/1000:.2f} kHz')

plt.legend()
    

plt.xlabel('R3 (kOhm)')
plt.ylabel('f (kHz)')
plt.ylim(6, 20)
plt.grid()
```
